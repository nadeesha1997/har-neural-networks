```python
label = LabelEncoder()
dataF_b['label'] = label.fit_transform(dataF_b['action'])

dataF_g['label'] = label.fit_transform(dataF_g['action'])

print(dataF_b.head())
# print(dataF_g.head())
```

#Standardized data#

```python
x = dataF_b[['attr_x','attr_y','attr_z']]
y = dataF_b['label']
x.shape
```

```python
scaler = StandardScaler()
x = scaler.fit_transform(x)


scaled_x = pd.DataFrame(data = x, columns=['x','y','z'])
scaled_x['label'] = y.values

scaled_x
```

#Frame preperation #

For accelerometer values only

```python
import scipy.stats as stats

frame_size = 200
hop_size = 50

def get_frames(df,frame_size, hop_size):
  N_FEATURES = 3

  frames = []
  labels = []

  for i in range(0,len(df) - frame_size,hop_size):
    x = df['x'].values[i: i + frame_size]
    y = df['y'].values[i: i + frame_size]
    z = df['z'].values[i: i + frame_size]

    label = stats.mode(df['label'][i: i + frame_size])[0][0]
    frames.append([x,y,z])
    labels.append(label)

  frames = np.asarray(frames).reshape(-1,frame_size,N_FEATURES)
  labels = np.asarray(labels)

  return frames, labels

```

```python
x,y = get_frames(scaled_x,frame_size,hop_size)
x.shape,y.shape
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0, stratify =y)
x_train.shape, x_test.shape
```

#Building LSTM model ***********

```python
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense, Flatten, LSTM
from keras.regularizers import l2
from tensorflow.keras.optimizers import Adam
```

```python
model=Sequential()
model.add(LSTM(32,return_sequences=True, input_shape = (frame_size, 3),
               kernel_regularizer = l2(0.000001), bias_regularizer = l2(0.000001), name = 'lstm_1'))

model.add(Flatten(name='flatten'))
model.add(Dense(64, activation='relu',kernel_regularizer = l2(0.000001), bias_regularizer = l2(0.000001), name='dense_1' ))
model.add(Dense(len(np.unique(y_train)), activation='softmax', 
                kernel_regularizer = l2(0.000001), bias_regularizer = l2(0.000001), name='output'))
model.summary()

```

```python
# compile 
model.compile(loss = 'sparse_categorical_crossentropy', optimizer=Adam(), metrics=['accuracy'])
```

```python
#prepare callbacks
from keras.callbacks import ModelCheckpoint

callbacks = [ModelCheckpoint('model.h5',save_weights_only=False, save_best_only=True, verbose=1)]
```

##start training

```python
history = model.fit(x_train,y_train, epochs=10, validation_data=(x_test,y_test), callbacks=callbacks)
```

```python
# stopping training here

# loading the best saved model 

from keras.models import load_model
model=load_model('model.h5')
model.summary()
```

```python
def plot_learning(history, epochs):
  epoch_range = range(1, epochs+1)

  plt.plot(epoch_range, history.history['accuracy'])
  plt.plot(epoch_range, history.history['val_accuracy'])
  plt.title('Model Accuracy')
  plt.legend(['Train','val'], loc='upper left')
  plt.show()

  plt.plot(epoch_range, history.history['loss'])
  plt.plot(epoch_range, history.history['val_loss'])
  plt.title('Model loss')
  plt.legend(['Train','val'], loc='upper left')
  plt.show()
```

```python
plot_learning(history, 10)
```

```python
from mlxtend.plotting import plot_confusion_matrix
from sklearn.metrics import confusion_matrix

predict_x=model.predict(x_test) 
y_pred=np.argmax(predict_x,axis=1)

mat = confusion_matrix(y_test,y_pred)
plot_confusion_matrix(conf_mat=mat, show_normed=True, figsize=(7,7))
```
